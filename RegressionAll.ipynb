## Notes:##

In L2 reguraization penalty term is added for weight coeficient hence optimization function becomes: 

#### min(||Y- X(θ)||^2 +λ||θ||^2)    ---Ridge regression
It will retain all of the features but will shrink the coefficients. Hence model will be complex
In L1 reguraization penalty term is added for weight coeficientoptimization function becomes: 

#### min(||Y- X(θ)||^2 +λ||θ||)   ---Lasso Regression

When we have correlated variables, it retains only one variable and sets other correlated variables to zero
As Laso regresssion does feature selection so may reduce the accuracy

Solution to above problem of lasso and ridge is elastic net regression, which is basically a hybrid of ridge and lasso regression. Its optimization function is

#### min(||Y- X(θ)||^2 +λ||θ||^2 + λ||θ||)   ---elastic net regression

```python
import math
RawFile="SparseData.csv"
ReadData = pd.read_csv(RawFile)
#SelectiveData=ReadData[['GPS_Latitude', 'GPS_Longitude',
#                        'CorrectedLatitude', 'CorrectedLongitude']]
SelectiveData=ReadData[['GPS_Latitude', 'GPS_Longitude','RawLatitude','RawLongitude',
                        'CorrectedLatitude', 'CorrectedLongitude']]
def angleFromCoordinate(lat1, long1,lat2,long2):
    
    dLon = (long2 - long1);
    y = math.sin(dLon) * math.cos(lat2);
    x = (math.cos(lat1) * math.sin(lat2) - math.sin(lat1) * math.cos(lat2)
         * math.cos(dLon))
    brng = math.atan2(y, x);
    brng = math.degrees(brng);
    brng = (brng + 360) % 360;
    brng = 360 - brng; # count degrees counter-clockwise - remove to make clockwise
    return brng;
ViewAngleRaw=[]
ViewAngleCorrect=[]
for index, row in SelectiveData.iterrows():   
    lat1= math.radians(row['GPS_Latitude'])
    long1= math.radians(row['GPS_Longitude'])
    lat2= math.radians(row['RawLatitude'])
    long2= math.radians(row['RawLongitude'])
    lat3= math.radians(row['CorrectedLatitude'])
    long3= math.radians(row['CorrectedLongitude'])
    angleRaw=angleFromCoordinate(lat1, long1,lat2,long2)
    angleCorrect=angleFromCoordinate(lat1, long1,lat3,long3)
    ViewAngleRaw.append(angleRaw)
    ViewAngleCorrect.append(angleCorrect)
AngleRawSeries= pd.Series(ViewAngleRaw)
AngleCorrectSeries= pd.Series(ViewAngleCorrect)

ReadData['ViewAngleRaw'] = pd.Series(AngleRawSeries,index=ReadData.index)
ReadData['ViewAngleCorrect'] = pd.Series(AngleCorrectSeries,index=ReadData.index)

```

```python
AngleDiff=[]
for index, row in ReadData.iterrows():
    angl1=row['ViewAngleCorrect']
    angl2=row['ViewAngleRaw']
    
    absAngle = angl1 - angl2
    if absAngle >180:
        absAngle=absAngle-360
    if absAngle < -180:
        absAngle= absAngle+360
    AngleDiff.append(absAngle)
AngleDiffSeries= pd.Series(AngleDiff)
ReadData['AngleDiff'] = pd.Series(AngleDiffSeries,index=ReadData.index)
ReadData['AngleDiffAbs']= ReadData['AngleDiff'].abs()
ReadData.to_csv("Viewangle%s" % RawFile, encoding='utf-8')
print("Angle result saved to the file viewangle" )
```
